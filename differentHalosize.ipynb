```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import AxesGrid
import xarray
import numpy as np
import logging

    # Ideally would use SSH, but some diag_tables only saved zos----for coupled waves
pp_root_halo2="/archive/Qian.Xiao/Qian.Xiao/FMS_Wave_Coupling_ZC/c5/postProcessing/NWA12.COBALT_all/19930101_results/"
pp_root_halo1="/archive/Qian.Xiao/Qian.Xiao/FMS_Wave_Coupling_ZC/postProcessing/NWA12.COBALT_all/19930101_results/"


ocean_daily_halo2= xarray.open_dataset(pp_root_halo2+"19930101.ocean_daily.nc")
ocean_daily_halo1= xarray.open_dataset(pp_root_halo1+"19930101.ocean_daily.nc")


model_grid_x = ocean_daily_halo2.variables['xh']
model_grid_y = ocean_daily_halo2.variables['yh']

#------------------------------------------ssh comparison between different momentum equations
ssh_halo2=ocean_daily_halo2.variables['ssh']
ssh_halo2_ave = ssh_halo2.mean('time')
t_halo2=ocean_daily_halo2.variables['time']
SST_halo2=ocean_daily_halo2.variables['SST']
SST_halo2_ave = SST_halo2.mean('time')
MLD_EN1_halo2=ocean_daily_halo2.variables['MLD_EN1']
MLD_EN1_halo2_ave = MLD_EN1_halo2.mean('time')
ssx_halo2=ocean_daily_halo2.variables['surface_stokes_x']
ssy_halo2=ocean_daily_halo2.variables['surface_stokes_y']
ssx_halo2_ave = ssx_halo2.mean('time')[:,1:]
ssy_halo2_ave = ssy_halo2.mean('time')[1:,:]

ssh_halo1=ocean_daily_halo1.variables['ssh']
ssh_halo1_ave = ssh_halo1.mean('time')
t_halo1=ocean_daily_halo1.variables['time']
SST_halo1=ocean_daily_halo1.variables['SST']
SST_halo1_ave = SST_halo1.mean('time')
MLD_EN1_halo1=ocean_daily_halo1.variables['MLD_EN1']
MLD_EN1_halo1_ave = MLD_EN1_halo1.mean('time')
ssx_halo1=ocean_daily_halo1.variables['surface_stokes_x']
ssy_halo1=ocean_daily_halo1.variables['surface_stokes_y']
ssx_halo1_ave = ssx_halo1.mean('time')[:,1:]
ssy_halo1_ave = ssy_halo1.mean('time')[1:,:]

dassh_halo2_ave=xarray.DataArray(ssh_halo2_ave, dims=['yh', 'xh'], coords={'xh': model_grid_x, 'yh': model_grid_y})
plt.figure()
dassh_halo2_ave.plot()
plt.title('Reference-wave(LT) with no ME modification with halosize=2')

###show the difference
diff_ssh_ave=ssh_halo1_ave - ssh_halo2_ave
da_diff_ssh=xarray.DataArray(diff_ssh_ave, dims=['yh', 'xh'], coords={'xh': model_grid_x, 'yh': model_grid_y})
plt.figure()
da_diff_ssh.plot()
plt.title('halosize=2 - halosize=1 ssh two-month mean')
plt.show()

daSST_halo2_ave=xarray.DataArray(SST_halo2_ave, dims=['yh', 'xh'], coords={'xh': model_grid_x, 'yh': model_grid_y})
plt.figure()
daSST_halo2_ave.plot()
plt.title('Reference-wave(LT) with no ME modification with halosize=2')

###show the difference
diff_SST_ave=SST_halo1_ave - SST_halo2_ave
da_diff_SST=xarray.DataArray(diff_SST_ave, dims=['yh', 'xh'], coords={'xh': model_grid_x, 'yh': model_grid_y})
plt.figure()
da_diff_SST.plot()
plt.title('halosize=2 - halosize=1 SST two-month mean')
plt.show()


daMLD_EN1_halo2_ave=xarray.DataArray(MLD_EN1_halo2_ave, dims=['yh', 'xh'], coords={'xh': model_grid_x, 'yh': model_grid_y})
plt.figure()
daMLD_EN1_halo2_ave.plot()
plt.title('Reference-wave(LT) with no ME modification with halosize=2')

###show the difference
diff_MLD_EN1_ave=MLD_EN1_halo1_ave - MLD_EN1_halo2_ave
da_diff_MLD_EN1=xarray.DataArray(diff_MLD_EN1_ave, dims=['yh', 'xh'], coords={'xh': model_grid_x, 'yh': model_grid_y})
plt.figure()
da_diff_MLD_EN1.plot(vmin=-10,vmax=10)
plt.title('halosize=2 - halosize=1 MLD_EN1 two-month mean')
plt.show()

dassx_halo2_ave=xarray.DataArray(ssx_halo2_ave, dims=['yh', 'xh'], coords={'xh': model_grid_x, 'yh': model_grid_y})
plt.figure()
dassx_halo2_ave.plot()
plt.title('Reference-wave(LT) with no ME modification with halosize=2')
###show the difference
diff_ssx_ave=ssx_halo1_ave - ssx_halo2_ave
da_diff_ssx=xarray.DataArray(diff_ssx_ave, dims=['yh', 'xh'], coords={'xh': model_grid_x, 'yh': model_grid_y})
plt.figure()
da_diff_ssx.plot()
plt.title('halosize=2 - halosize=1 ssx two-month mean')
plt.show()

dassy_halo2_ave=xarray.DataArray(ssy_halo2_ave, dims=['yh', 'xh'], coords={'xh': model_grid_x, 'yh': model_grid_y})
plt.figure()
dassy_halo2_ave.plot()
plt.title('Reference-wave(LT) with no ME modification with halosize=2')
###show the difference
diff_ssy_ave=ssy_halo1_ave - ssy_halo2_ave
da_diff_ssy=xarray.DataArray(diff_ssy_ave, dims=['yh', 'xh'], coords={'xh': model_grid_x, 'yh': model_grid_y})
plt.figure()
da_diff_ssy.plot()
plt.title('halosize=2 - halosize=1 ssy two-month mean')
plt.show()
```
